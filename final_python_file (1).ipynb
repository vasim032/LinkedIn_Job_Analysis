# Importing all required libraries

```python
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
```

# Connecting webdriver

```python
s=Service('chromedriver.exe')
browser = webdriver.Chrome(service=s)
```

```python
link = "https://www.linkedin.com/"
browser.get(link)
```

# Moving to Job collections LInk

```python
browser.get("https://www.linkedin.com/jobs/collections/")
```

# List container for storing useful html code

```python
#table 1
designation = []
location = []
job_link = []

#table 2
name = []
industry = []
emp_count = []
linkedin_followers = []

#table 3
applicants = []
involvement = []
work_type = []
```

# code for extracting useful data to store in lists


```python
for i in range(1,41):
    #button path for pages
    path ='//button[@aria-label="Page {}"]'.format(i)
    
    #button clicking    
    browser.find_element(By.XPATH, path).click()
    
#     # rest time for button clicking
#     time.sleep(1) 
    
    
    
    
    #html data
    src = browser.page_source
    soup = BeautifulSoup(src, 'lxml')
    
    #main page of one job data
    lk=soup.findAll(class_="disabled ember-view job-card-container__link")
    
    
    #link of a single job data
    for i in lk:
        # links
        li=i['href']
        
        
        #every page data
        every_page =BeautifulSoup(browser.page_source,'lxml')
        
           
        
#         # rest time for moving next window
#         time.sleep(1)
        
        #movig to link using next window_of_ chrome
        browser.switch_to.new_window('tab')
        job_link.append("https://www.linkedin.com{}".format(li))
        browser.get("https://www.linkedin.com{}".format(li))
        
#          # rest time for Waiting at next window
#         time.sleep(1)
        
         #finding company name
        try:
            c_name = browser.find_elements(By.CLASS_NAME,'jobs-unified-top-card__company-name')
            name.append(c_name[0].text) 
        except:
            name.append("not specify")
         
        #finding designation   
        try:
            desig = browser.find_elements(By.CLASS_NAME,'jobs-unified-top-card__job-title')
            designation.append(desig[0].text)
        except:
            designation.append("not specify")
        
        #finding applicants
        try:
            applic = browser.find_elements(By.XPATH,'/html/body/div[5]/div[3]/div/div[1]/div[1]/div/div[1]/div/div/div[1]/div[1]/span[2]/span[2]/span')
            applicants.append(applic[0].text) 
        except:
            applicants.append("0") 
            
            
            
        #finding work type
        try:
            work_t = browser.find_elements(By.CLASS_NAME,'jobs-unified-top-card__workplace-type')
            work_type.append(work_t[0].text)  
        except:
            work_type.append("not specify")
        
        
        #finding involvement
        try:
            invo = browser.find_elements(By.CLASS_NAME,'jobs-unified-top-card__job-insight')
            involvement.append(invo[0].text)
        except:
            involvement.append("not specify")
         
        #finding employ count
        try:
            emp = browser.find_elements(By.CLASS_NAME,'jobs-unified-top-card__job-insight')
            emp_count.append(emp[1].text)
        except:
            emp_count.append("not specify")
        
        #finding location
        try:
            loc = browser.find_elements(By.CLASS_NAME,'jobs-unified-top-card__bullet')
            location.append(loc[0].text)
        except:
            location.append("not specify")
    
        
        
        
        #every page data
        every_page =BeautifulSoup(browser.page_source,'lxml')
        
        
        # details page//last page
        s = []
        src = browser.page_source
        soup = BeautifulSoup(src, 'lxml')
        detai = soup.findAll(class_='ember-view t-black t-normal')
        for u in detai:
            s.append(u)
        
        
            
        
       
    
        # for table 2-3 link 
        for i in s:
            pr = i['href']
            
            
#             # rest time for moving next window
#             time.sleep(1)
            
            
            #movig to link using next window_of_ chrome
            browser.switch_to.new_window('tab')
            browser.get("https://www.linkedin.com{}".format(pr))
            
            # rest time for Waiting at next window
            time.sleep(6)
            
            #finding industry
            try:
                ind = browser.find_elements(By.CLASS_NAME,'org-top-card-summary-info-list__info-item')
                industry.append(ind[0].text)
            except:
                industry.append("not specify")

            #finding followers    
            try:
                follow = browser.find_elements(By.XPATH,'//*[@id="ember28"]/div[2]/div[1]/div[1]/div[2]/div/div/div[2]/div[2]')
                linkedin_followers.append(follow[0].text)
            except:
                linkedin_followers.append("N/A")
                
                
            # current window close        
            browser.close()
            #switch to main(starting) tab/window
            browser.switch_to.window(browser.window_handles[-1])
     
        
        

        
              
        # current window close        
        browser.close()
        #switch to main tab/window
        browser.switch_to.window(browser.window_handles[0])
        
```
